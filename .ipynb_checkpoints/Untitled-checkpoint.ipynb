This is an implementation of Naive Bayes on the Dataset.

```python
#This is a step to check if there are null values in the dataset
df.isnull().sum()
```

```python

#This step is to standardize data so that each feature is on the same scale
num_features = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology',
                'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm',
                'Horizontal_Distance_To_Fire_Points', 'Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3',
                'Wilderness_Area4','Soil_Type1','Soil_Type2','Soil_Type3','Soil_Type4','Soil_Type5','Soil_Type6',
               'Soil_Type7','Soil_Type8','Soil_Type9','Soil_Type10','Soil_Type11','Soil_Type12','Soil_Type13','Soil_Type14',
               'Soil_Type15','Soil_Type16','Soil_Type17','Soil_Type18','Soil_Type19','Soil_Type20','Soil_Type21','Soil_Type22',
               'Soil_Type23','Soil_Type24','Soil_Type25','Soil_Type26','Soil_Type27','Soil_Type28','Soil_Type29','Soil_Type30',
               'Soil_Type31','Soil_Type32','Soil_Type33','Soil_Type34','Soil_Type35','Soil_Type36','Soil_Type37','Soil_Type38',
               'Soil_Type39','Soil_Type40']

scaled_features = {}
for each in num_features:
    mean, std = df[each].mean(), df[each].std()
    scaled_features[each] = [mean, std]
    df.loc[:, each] = (df[each] - mean)/std
    
#df = pd.read_csv("covtype.csv")
    
print(df)
```

```python
features = df.values[:,:54]
target = df.values[:,54]
features_train, features_test, target_train, target_test = train_test_split(features, target,test_size = 0.33, random_state = 10)
print(target)
```

```python
clf = GaussianNB()
clf.fit(features_train, target_train)
target_predict = clf.predict(features_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(target_test, target_predict, normalize = True)
```

    This step is to try the SGDClassifier

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss = "hinge", max_iter = 5)
clf.fit(features_train, target_train)
```

```python
target_predict = clf.predict(features_test)
accuracy_score(target_test, target_predict, normalize = True)
```

```python
from sklearn.kernel_approximation import RBFSampler
rbf_feature = RBFSampler(gamma=1, random_state=1)
X_features = rbf_feature.fit_transform(features_train)
X_test = rbf_feature.transform(features_test)
print(X_features.shape)
```

```python
clf = SGDClassifier(max_iter=5)
clf.fit(X_features, target_train) 
clf.score(X_test, target_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=5,n_jobs=-1)  
classifier.fit(features_train, target_train)
target_predict = classifier.predict(features_test)
accuracy_score(target_test, target_predict, normalize = True)
```

```python
#Linear Regression without standardization
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(features_train, target_train)
reg.score(features_test, target_test)
```
